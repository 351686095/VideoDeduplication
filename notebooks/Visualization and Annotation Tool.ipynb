## Annotation Tool

Objective: 

Allow seamless interaction with the matches report.

Features:

    1) Video visualization through frame strips (with scroll)
    2) Side by side comparison between query and matched video
    3) Real-time annotation
    4) Annotation summarization
    
Can be adapted to run with the template matching notebook

```python
# Manually set the path to the frame_level foldere
FRAME_LEVEL_SAVE_FOLDER = 'data/representations/frame_level'
REPORT_PATH = 'data/matches_at_0.75_distance_filtered.csv'
```

```python
annotator = Annotator(REPORT_PATH,
                      'annotated_report.csv',
                       annotation_label='annotation_is_match',
                       frames_directory= FRAME_LEVEL_SAVE_FOLDER)
```

## Matches Review / Annotation Tool

```python
annotator.run()
```

```python
annotator.summary()
```
